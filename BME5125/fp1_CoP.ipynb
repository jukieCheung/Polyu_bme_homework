```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/mnt/e/BME5150/g11t3_oppo_fp(1).xlsx'
df = pd.DataFrame(pd.read_excel(path, header=2))
```

```python
df[2:1878]
```

```python
df1 = df[['FP1:3204 - Force','FP1:3204 - Force.1','FP1:3204 - Force.2']]
df11 = df[['FP1:3204 - Moment','FP1:3204 - Moment.1','FP1:3204 - Moment.2']]
df111 = df[['FP1:3204 - CoP','FP1:3204 - CoP.1','FP1:3204 - CoP.2']]
```

```python
df2 = df1[2:1878]
df21 = df11[2:1878]
df211 = df111[2:1878]
```

```python
# 生成 x 轴数据
x = [i * 0.001 for i in range(0, len(df) - 2)]
```

```python
# 绘制图表
plt.figure(figsize=(10, 6))  # 设置图形大小
for column in df2.columns:
    plt.plot(x, df2[column], label=column)
#for column in df21.columns:
#    plt.plot(x, df21[column], label=column)
for column in df211.columns:
    plt.plot(x, df211[column], label=column)

# 添加标题和标签
plt.title('Example Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()  # 添加图例
plt.grid(True)  # 添加网格线
plt.show()
```

```python
# 采样间隔时间
sampling_interval = 0.001
start_index = int(0.75 / 0.001)
end_index = int(1.4 / 0.001)
df211 = df211[start_index:end_index]
# 计算速度
velocity_df = df211.diff(periods=2) / (2 * sampling_interval)

# 重新设置索引
velocity_df.index = range(1, len(velocity_df) + 1)
x2 = [i * 0.001 for i in range(0, len(df211))]  # 0.004 的采样率
# 绘制图表
plt.figure(figsize=(10, 6))  # 设置图形大小
#for column1 in df01.columns:
#    plt.plot(x2, df01[column1], label=column1, linestyle='--')

for column in velocity_df.columns:
    plt.plot(velocity_df.index * sampling_interval, velocity_df[column], label=(column+'velocity'))

# 添加标题和标签
plt.title('Velocity Plot')
plt.xlabel('Time (s)')
plt.ylabel('Velocity')
plt.legend()  # 添加图例
plt.grid(True)  # 添加网格线
plt.show()
```
