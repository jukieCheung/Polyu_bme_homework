```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
path = '/mnt/e/BME5150/g11t3_oppo_fp(1).xlsx'
df = pd.read_excel(path, header=2)

# 选择需要计算受力总和的列，并删除前两行（单位信息）
df1 = df[['FP1:3204 - Force','FP1:3204 - Force.1','FP1:3204 - Force.2']]
df2 = df1.iloc[2:1878]  # 切片选取需要的时间范围内的数据

# 计算 x 轴数据
x = [i * 0.001 for i in range(0, len(df2))]

# 绘制图表
plt.figure(figsize=(10, 6))  # 设置图形大小
for column in df2.columns:
    plt.plot(x, df2[column], label=column)

# 确定时间范围内的索引
start_index = int(0.75 / 0.001)
end_index = int(1.4 / 0.001)

# 计算时间范围内的受力总和
#df3 = df2.iloc[start_index:end_index+1]
df3 = df2[['FP1:3204 - Force.1']]
total_force = df3.iloc[start_index:end_index+1].sum().sum()
print("0.75s到1.4s的受力总和为:", total_force)

# 添加标题和标签
plt.title('Example Plot')
plt.xlabel('Time (s)')
plt.ylabel('Force')
plt.legend()  # 添加图例
plt.grid(True)  # 添加网格线
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
path = '/mnt/e/BME5150/g11t3_same_fp.xlsx'
df = pd.read_excel(path, header=2)

# 选择需要计算受力总和的列，并删除前两行（单位信息）
df1 = df[['FP1:3204 - Force','FP1:3204 - Force.1','FP1:3204 - Force.2']]
df2 = df1.iloc[2:1878]  # 切片选取需要的时间范围内的数据

# 计算 x 轴数据
x = [i * 0.001 for i in range(0, len(df2))]

# 绘制图表
plt.figure(figsize=(10, 6))  # 设置图形大小
for column in df2.columns:
    plt.plot(x, df2[column], label=column)

# 确定时间范围内的索引
start_index = int(0.5 / 0.001)
end_index = int(0.95 / 0.001)

# 计算时间范围内的受力总和
#df3 = df2.iloc[start_index:end_index+1]
df3 = df2[['FP1:3204 - Force.1']]
total_force = df3.iloc[start_index:end_index+1].sum().sum()
print("0.5s到0.95s的受力总和为:", total_force)

# 添加标题和标签
plt.title('Example Plot')
plt.xlabel('Time (s)')
plt.ylabel('Force')
plt.legend()  # 添加图例
plt.grid(True)  # 添加网格线
plt.show()
```

```python
path = '/mnt/e/BME5150/g11t3_oppo_fp(1).xlsx'
df = pd.DataFrame(pd.read_excel(path, header=2))
df1 = df[['FP1:3204 - Force','FP1:3204 - Force.1','FP1:3204 - Force.2']]
df2 = df1[2:1878]
# 生成 x 轴数据
x1 = [i * 0.001 for i in range(0, len(df) - 2)]

path1 = '/mnt/e/BME5150/g11t3_same_fp.xlsx'
df0 = pd.DataFrame(pd.read_excel(path1, header=2))
df01 = df0[['FP1:3204 - Force','FP1:3204 - Force.1','FP1:3204 - Force.2']]

df02 = df01.drop([0, 1])
x2 = [i * 0.001 for i in range(0, len(df0) - 2)]

# 绘制图表
plt.figure(figsize=(10, 6))  # 设置图形大小

# 绘制第一个 DataFrame 的数据
for column in df1.columns:
    plt.plot(x1, df2[column], label=column)

# 绘制第二个 DataFrame 的数据
for column1 in df01.columns:
    plt.plot(x2, df02[column1], label=column1, linestyle='--')

# 添加标题和标签
plt.title('Combined Plot')
plt.xlabel('Time')
plt.ylabel('Data')
plt.legend()  # 添加图例
plt.grid(True)  # 添加网格线
plt.show()
```
