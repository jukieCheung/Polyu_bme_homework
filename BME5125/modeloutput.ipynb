```python
import pandas as pd
import matplotlib.pyplot as plt

# 
path1 = '/mnt/e/BME5150/g11t3_oppo_modeloutput.xlsx'
df0 = pd.read_excel(path1, header=2)

# 
df01 = df0[['G11:CentreOfMass','G11:CentreOfMass.1','G11:CentreOfMass.2']]
df01 = df01.drop([0, 1])

# 
sampling_interval = 0.004

# 
velocity_df = df01.diff(periods=2) / (2 * sampling_interval)

# 
velocity_df.index = range(1, len(velocity_df) + 1)
x2 = [i * 0.004 for i in range(0, len(df01))]  # 0.004 
# 
plt.figure(figsize=(10, 6))  # 
for column1 in df01.columns:
    plt.plot(x2, df01[column1], label=column1, linestyle='--')

for column in velocity_df.columns:
    plt.plot(velocity_df.index * sampling_interval, velocity_df[column], label=(column+'velocity'))

# 
plt.title('Velocity Plot')
plt.xlabel('Time (s)')
plt.ylabel('Velocity')
plt.legend()  # 
plt.grid(True)  # 
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 
path1 = '/mnt/e/BME5150/g11t3_oppo_modeloutput.xlsx'
df0 = pd.read_excel(path1, header=2)

# 
#df01 = df0[['G11:RShoulderForce','G11:RShoulderForce.1','G11:RShoulderForce.2','G11:LShoulderForce','G11:LShoulderForce.1','G11:LShoulderForce.2']]
df01 = df0[['G11:RShoulderForce.1','G11:RShoulderForce.2','G11:LShoulderForce.1','G11:LShoulderForce.2']]
df01 = df01.drop([0, 1])

# 
sampling_interval = 0.004

# 
momentum_df = pd.DataFrame()
for column in df01.columns:
    force = df01[column].values
    momentum = np.zeros_like(force)
    for i in range(1, len(force)):
        momentum[i] = force[i] * sampling_interval / 2 + force[i-1] * sampling_interval + force[i-1] * sampling_interval / 2
    momentum_df[column] = momentum

# 
momentum_df.index = range(1, len(momentum_df) + 1)

# 
plt.figure(figsize=(10, 6))  # 
for column in momentum_df.columns:
    plt.plot(momentum_df.index * sampling_interval, momentum_df[column], label=column, linestyle='-.')

# 
plt.title('Momentum Plot')
plt.xlabel('Time (s)')
plt.ylabel('Momentum')
plt.legend()  # 
plt.grid(True)  # 
plt.show()
```
