```python
import pandas as pd
import matplotlib.pyplot as plt

path = '/mnt/e/BME5150/g11t3_oppo_fp(1).xlsx'
df = pd.DataFrame(pd.read_excel(path, header=2))
df1 = df[['FP1:3204 - Force','FP1:3204 - Force.1','FP1:3204 - Force.2']]
df2 = df1[2:1878]
#  x 
x1 = [i * 0.001 for i in range(0, len(df) - 2)]

path1 = '/mnt/e/BME5150/g11t3_oppo_trajectories.xlsx'
df0 = pd.DataFrame(pd.read_excel(path1, header=2))
df01 = df0[['G11:RBAK','G11:RBAK.1','G11:RBAK.2']]
df01 = df01.drop([0, 1])

# 
sampling_interval = 0.004

# 
velocity_df = df01.diff(periods=2) / (2 * sampling_interval)

# 
velocity_df.index = range(1, len(velocity_df) + 1)
# 
plt.figure(figsize=(10, 6))  # 

#  DataFrame 
for column in df1.columns:
    plt.plot(x1, df2[column], label=column)

for column in velocity_df.columns:
    plt.plot(velocity_df.index * sampling_interval, velocity_df[column], label=(column+'velocity'), linestyle='--')

# 
plt.title('Velocity Plot')
plt.xlabel('Time (s)')
plt.ylabel('Velocity')
plt.legend()  # 
plt.grid(True)  # 
plt.show()
```
