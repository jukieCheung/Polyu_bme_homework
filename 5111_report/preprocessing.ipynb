```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
resting_data = pd.read_csv('resting_data.csv')
walking_data = pd.read_csv('walking_data.csv')
horizontal_jumping = pd.read_csv('horizontal_jumping.csv')
vertical_jumping = pd.read_csv('vertical_jumping.csv')
```

```python
# 
data = vertical_jumping

# 
points_per_2_seconds = int(2.0 / (data["Time (s)"].iloc[1] - data["Time (s)"].iloc[0]))

# 
segmented_data = []

# 
start_index = 0

# 
for end_index in range(points_per_2_seconds, len(data), points_per_2_seconds):
    segment = data[start_index:end_index]
    segmented_data.append(segment)
    start_index = end_index

# 
segment = data[start_index:]
segmented_data.append(segment)

# segmented_data 
```

```python
min_x_values = []
max_x_values = []
mean_x_values = []
peak_to_peak_x_values = []

min_y_values = []
max_y_values = []
mean_y_values = []
peak_to_peak_y_values = []

min_z_values = []
max_z_values = []
mean_z_values = []
peak_to_peak_z_values = []

for segment in segmented_data[1:26]:
    min_x = segment["Linear Acceleration x (m/s^2)"].min()
    max_x = segment["Linear Acceleration x (m/s^2)"].max()
    mean_x = segment["Linear Acceleration x (m/s^2)"].mean()
    peak_to_peak_x = max_x - min_x

    min_y = segment["Linear Acceleration y (m/s^2)"].min()
    max_y = segment["Linear Acceleration y (m/s^2)"].max()
    mean_y = segment["Linear Acceleration y (m/s^2)"].mean()
    peak_to_peak_y = max_y - min_y

    min_z = segment["Linear Acceleration z (m/s^2)"].min()
    max_z = segment["Linear Acceleration z (m/s^2)"].max()
    mean_z = segment["Linear Acceleration z (m/s^2)"].mean()
    peak_to_peak_z = max_z - min_z

    min_x_values.append(min_x)
    max_x_values.append(max_x)
    mean_x_values.append(mean_x)
    peak_to_peak_x_values.append(peak_to_peak_x)

    min_y_values.append(min_y)
    max_y_values.append(max_y)
    mean_y_values.append(mean_y)
    peak_to_peak_y_values.append(peak_to_peak_y)

    min_z_values.append(min_z)
    max_z_values.append(max_z)
    mean_z_values.append(mean_z)
    peak_to_peak_z_values.append(peak_to_peak_z)

```

```python
output_feature = pd.DataFrame({'min_x':min_x_values, 'max_x':max_x_values, 'mean_x':mean_x_values, 'peak_to_peak_x':peak_to_peak_x_values,
                               'min_y':min_y_values, 'max_y':max_y_values, 'mean_y':mean_y_values, 'peak_to_peak_y':peak_to_peak_y_values,
                               'min_z':min_x_values, 'max_z':max_x_values, 'mean_z':mean_x_values, 'peak_to_peak_z':peak_to_peak_x_values,
                               })

output_feature.to_csv('vertical_feature.csv', index=None, encoding='utf8')
```
